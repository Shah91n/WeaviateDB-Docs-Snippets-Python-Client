```python
from dotenv import load_dotenv
import os
from engram import EngramClient

# Load environment variables from .env file
load_dotenv(override=True)

client = EngramClient(api_key=os.getenv("ENGRAM_API_KEY"))
print("Engram client initialized")
```

```python
# Example: client.memories.add()
run = client.memories.add(
    "Incidents had no owner at the start. We added an incident lead to improve SLA response time.",
    user_id="demo_user_1",
)
print("Run ID:", run.run_id)
```

```python
# Example: client.runs.get()
status_now = client.runs.get(run.run_id)
print("Current status:", status_now.status)
```

```python
# Example: client.runs.wait()
status = client.runs.wait(run.run_id, timeout=30)
print("Status:", status.status)
```

```python
# Example: client.memories.search()
results = client.memories.search(
    query="incident owner and SLA",
    user_id="demo_user_1",
)
for memory in results:
    print(memory.id, "->", memory.content)
```

```python
# Example: client.memories.get()
if len(results) > 0:
    first_id = results[0].id
    memory = client.memories.get(first_id, user_id="demo_user_1")
    print("Loaded memory:", memory.content)
else:
    print("No memory found. Run add/search first.")
```

```python
# Example: client.memories.delete()
if len(results) > 0:
    memory_to_delete = results[0].id
    client.memories.delete(memory_to_delete, user_id="demo_user_1")
    print("Deleted memory:", memory_to_delete)
else:
    print("No memory to delete. Run add/search first.")
```
